## Clipping

**CLipping** Meanss using a polygon or polygons to slelcet only geospatila data within them.
This is simple with a `geopandas.GeoDataFrame` using the `clip()` function:

```python
updated_geo_df = geopandas.clip(geodf, mask)
```

where
- `updated_geodf` = output of the method, the **intersection** of the gepometries in `geodf` with the `mask`
- `geodf` = `gpd.geodf` we want to clip
- `mask` = `gpd.geodf` witht the polygons we want to use for clipping. The `mask`  must be in the same CRS as `geo_df`!

````python
ak_places = gpd.clip(places, alaska)

fig, ax = plt.subplots()

alaska.plot(ax = ax)
places.plot(ax = ax, color = "red")
````

## Clipping with a bounding box

We will clip the `usa_roads` geo-dataframe witht he  **bounding_box** of `alaska` instead of it's polygons.

To create a bounding box, we will use the `box()` functoin from `shapely.geometry`. The syntax for the box is:

```python
box(minx, minz, maxx, maxy)
```

The output is a polygon representing a box constructed like this:

SHOW DIRAGRAM:

If we want to create a `shapely` polygon from the bounds of a geo-df `gdf` we can use this syntax:

```python
box(*gdf.total_bounds)
```

````python
bbox = box(*alaska.total_bounds)
print(type(bbox))
bbox
````

### * = unpacking operator

In the last suntax we use an asterisk `*` as an **unpacking operator** on the array `alaska.total_bounds`. Think about it as unpacking the elements of `alaska.total_bounds` and passing them one-by-one to the parameters `minx`, `miny`, `maxx`, `maxy` of the `box()` function.

This is a good article explaining more about unpacking with * in Python: https://geekflare.com/python-unpacking-operators/

````python
# CReate a geo_dataframe form the bounding box
ak_bbox = gpd.GeoDataFrame(geometry = [bbox], # Assign geometry column
                          crs = alaska.crs # Assign a CRS
                          )

print(type(ak_bbox))
ak_bbox
````

Now we can clip the roads using the alaska bounding box

````python
# Clip roads using ak_bbox
ak_complete_roads = gpd.clip(usa_roads.to_crs(ak_bbox.crs), ak_bbox)
````

````python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize = (10, 10))

ak_roads.plot(ax = ax1)
ax1.set_title('Roads clipped with AK multipolygon')

ak_complete_roads.plot(ax = ax2)
ax2.set_title('Roads clipped with AK bounding box')
plt.show()
````
