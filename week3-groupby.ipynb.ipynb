```python
import numpy as np
import pandas as pd

# load the data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)
```

```python
# Maximum body mass by year and species
(penguins.groupby(['year', 'species'])
          .body_mass_g
          .max()
          .rename('max_body_mass_g')
)
```

```python
(penguins.groupby(['year', 'species'])
          .body_mass_g
          .max()
          .sort_values()
          .plot(kind = 'barh',
                  title = 'Penguin body mass by species and year',
                  xlabel = ('Year', 'Species'))
          .set_xlabel(xlabel = 'Maximum body mass (g)')
)
```
